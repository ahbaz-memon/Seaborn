```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')#styling for beauty
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.relplot(x='total_bill',
            y='tip',
            data=data,
            kind='scatter')
```

```python
sns.relplot(x='total_bill',
            y='tip',
            data=data,
            kind='scatter',
            style='smoker')
```

```python
 sns.relplot(x='total_bill',
            y='tip',
            data=data,
            kind='scatter',
            hue='sex',
            style='smoker')
```

```python
 sns.relplot(x='total_bill',
            y='tip',
            data=data,
            kind='scatter',
            hue='sex',
            style='smoker',
            size='size')#can increase size by sizes=(min,max)
```

```python
sns.scatterplot(x='total_bill',
            y='tip',
            data=data,
            hue='sex',
            style='smoker',
            size='size')
```
