```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

### 1. categorical scatterplots

##### a. strip-plot 

```python
sns.catplot(x='day',y='tip',data=data,kind='strip')
#can also use -> sns.stripplot(x='day',y='tip',data=data)
```

```python
sns.catplot(x='day',
            y='tip',
            data=data,
            kind='strip',
            jitter=2 #controls noise with same level data 
           )
```

##### b. swarmplot

```python
sns.catplot(x='day',
            y='tip',
            data=data,
            kind='swarm'            
           )
# can also use sns.swarmplot(x='day',y='tip',data=data)
```

```python
sns.catplot(x='day',
            y='tip',
            data=data,
            kind='swarm',
            hue='sex'       
           )
```

### 2. categorical distribution plots

##### a. box-plot 

```python
sns.boxplot(data['tip'])
```

```python
sns.catplot(x='day',
            y='total_bill',
            data=data,
            kind='box'
           )
```

```python
sns.catplot(x='day',
            y='total_bill',
            data=data,
            kind='box',
            hue='sex'
            )
```

##### b. violin-plot(boxplot + PDF)

```python
sns.violinplot(data['tip'])
```

```python
sns.violinplot(data['total_bill'])
```

```python
sns.catplot(x='day',
            y='total_bill',
            data=data,
            kind='violin',
           )
```

```python
sns.catplot(x='day',
            y='total_bill',
            data=data,
            kind='violin',
            hue='sex'
           )
```

```python
sns.catplot(x='day',
            y='total_bill',
            data=data,
            kind='violin',
            hue='sex',
            split=True
           )
```

### 3. categorical estimate plots

##### a. Point-plot 

```python
sns.pointplot(data['total_bill'])
```

```python
sns.catplot(x='day',
            y='total_bill',
            data=data,
            kind='point'
           )
```

```python
sns.catplot(x='day',
            y='total_bill',
            data=data,
            kind='point',
            hue='sex'
           )
```

##### b. bar-plot

```python
sns.barplot(x='smoker',
            y='total_bill',
            data=data,
           )
```

```python
sns.catplot(x='smoker',
            y='total_bill',
            data=data,
            kind='bar',
            hue='sex'
           )
```

```python
sns.catplot(x='smoker',
            y='total_bill',
            data=data,
            kind='bar',
            hue='sex',
            estimator=np.var #default value is np.value
           )
```

##### c. count-plot ~ histogram

```python
sns.countplot(data['sex'])
```

```python
sns.catplot(x='sex',
              data=data,
              kind='count',
              hue='smoker'
             )
```

### 4. categorical categorical plots

##### a. Heat-maps

```python
flights=sns.load_dataset('flights')
```

```python
flights.shape
```

```python
flights.head()
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False) #color gets off
```

```python
sns.heatmap(x,cbar=True,linewidths=2) #line width within color cell
```

```python
sns.heatmap(x,
            cbar=True,
            linewidths=2,
            linecolor='m'
           ) #line color in width replace within color cell
```

```python
sns.heatmap(x,
            cbar=True,
            linewidths=2.5,
            linecolor='k',
            annot=True
           )
```

```python
sns.heatmap(x,
            cbar=True,
            linewidths=2,
            linecolor='k',
            annot=True,
            fmt='d'
           )
```

```python
sns.heatmap(x,
            cbar=True,
            linewidths=2,
            cmap='viridis'
           )
```

```python
sns.heatmap(x,
            cbar=True,
            linewidths=2,
            cmap='summer'
           )
```

```python
sns.heatmap(x,
            cbar=True,
            linewidths=2,
            cmap='winter'
           )
```

##### b. cluster-maps

```python
sns.clustermap(x)
```

```python
#all parameter used in heatmap is similar
# and for clustering it uses metric : euclidian dist
```

```python
sns.clustermap(x,
              metric='correlation'
              )
```

```python
sns.clustermap(x,
               metric='correlation',
               z_score=0,
               annot=True
              )
```

```python
sns.clustermap(x,
               metric='correlation',
               z_score=0,
               annot=True,
               row_cluster=False# similar to colomn cluster
              )
```
